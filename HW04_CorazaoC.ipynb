# 4.1 Merge df_canton_all, df_canton, and df_census into one dataset.
OBJECTIVE: One of the issues with the data that was used for regression was that the weeks
that did not have any crimes were not included. See the table below, for each of the
Subzones, there should be 516 weeks of data. However, in the CantonPoliceDept_HW04
dataset, some are less than 516. That is due to those subzones not having any calls or
dispositions in some weeks. But we would still want to include those weeks for classification
modeling.
• Therefore, for the merge, we want to keep ALL of the records from df_canton_all and
merge the ones that are common from df_canton. We also want df_census data for all
records in df_canton_all. Name the DataFrame df_all.

```python
# CodeBlock SubSection 4.1
# merging with left join to get all records of df_canton_all and common records from df_canton
df_all = pd.merge(df_canton_all, df_canton, how='left', on='SUB_YEAR_WEEK')
df_all.info()
```

```python
# CodeBlock SubSection 4.1
# merging outer to get all records from df_all and df_census
df_all = pd.merge(df_all, df_census, how='outer', on='Subzone')
df_all.info()
```

```python
# CodeBlock SubSection 4.1
# checking data
df_all.head()
```

```python
# CodeBlock SubSection 4.1
# filling nulls with 0
df_all = df_all.fillna(0)
df_all
```
